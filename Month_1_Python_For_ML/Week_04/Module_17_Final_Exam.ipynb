```python
dept_wise_median= df.groupby('Department')['Salary'].median()
dept_wise_median
```

```python
df['Salary'] = df['Salary'].fillna(df['Department'].map(dept_wise_median))
df
```

```python
print(df['Salary'].isna().sum())
```

```python
updated= df[(df['ExperienceYears']>12)& (df['ProjectCount']>5)]
total_salary= updated['Salary'].sum()
total_salary
```

```python
filter_employee= df[((df['Department'] == 'IT') | (df['Department'] == 'Finance'))
& (df['Age'] >= 30) & (df['Age'] <= 45) & (df['PerformanceScore'] > 88)
]
sorted=filter_employee.sort_values(by='Salary', ascending=False)
final_output=sorted[['Name', 'Department', 'Salary', 'PerformanceScore']]
final_output
```
