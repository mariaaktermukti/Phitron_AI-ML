# Module 1 ->  Descriptive Statistics and Distributions


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#part A -> Ques 1

marks=[ 50, 60, 65, 70, 75, 80, 85, 90]
df= pd.DataFrame({"Marks": marks})
df
```

```python
#Find mean, median, mode.

mean_marks= df['Marks'].mean()
print("Mean is =", mean_marks)

median_marks= df['Marks'].median()
print("Median is =", median_marks)

#find mode value
mode_value = df['Marks'].mode()
if len(mode_value) == len(df):
    print("Mode: No unique mode")
else:
    print("Mode is =", list(mode_value))
```

```python
# Replace 90 with 900 and recompute mean. Explain the effect of outliers.
import pandas as pd

new_marks = pd.DataFrame({'Marks': [50, 60, 65, 70, 75, 80, 85, 900]})

new_marks
```

```python
# mean
new_mean = new_marks['Marks'].mean()
print("Mean is =", new_mean)

new_median= new_marks['Marks'].median()
print("Median is =", new_median)
```

```python
#histrogram diye visualization
sns.histplot(df['Marks'], kde=True, color='skyblue', label='Original Data')
sns.histplot(new_marks['Marks'], kde=True, color='orange', label='With Outlier')


plt.axvline(mean_marks, color='blue', linestyle='--', label=f'Original Mean = {mean_marks:.2f}')
plt.axvline(new_mean, color='red', linestyle='--', label=f'Outlier Mean = {new_mean:.2f}')

# plt.legend()
plt.title('Effect of Outlier on Marks')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Type'] = 'Original'
new_marks['Type'] = 'With Outlier'
combined = pd.concat([df, new_marks])

#Plot
plt.figure(figsize=(8,5))
sns.histplot(data=combined, x='Marks', hue='Type', kde=True, palette=['skyblue','orange'], alpha=0.5)

# Add mean lines
plt.axvline(mean_marks, color='blue', linestyle='--', label=f'Original Mean = {mean_marks:.2f}')
plt.axvline(new_mean, color='red', linestyle='--', label=f'With Outlier Mean = {new_mean:.2f}')

# Focus x-axis only up to 150 to see the effect clearly
plt.xlim(40, max(new_mean + 50, 150))  # Extend x-axis to show the new_mean
plt.legend()
plt.title('Effect of Outlier on Mean')
plt.xlabel('Marks')
plt.ylabel('Frequency')

plt.show()
```

# Part B  Percentiles, IQR and Z-Score

```python
#task 3

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Cat weights
weights = np.array([2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5])
weights
```

```python
# Calculate percentiles
P25 = np.percentile(weights, 25)
P50 = np.percentile(weights, 50)
P75 = np.percentile(weights, 75)
IQR = P75 - P25

print(f"P25 = {P25}")
print(f"P50 = {P50}")
print(f"P75 = {P75}")
print(f"IQR = {IQR}")
```

```python
sns.boxplot(weights, color='yellow')
plt.title("Cat Weights Boxplot")
plt.xlabel("Weight (kg)")
plt.show()
```

```python
# Calculate outlier bounds
lower_bound = P25 - 1.5 * IQR
upper_bound = P75 + 1.5 * IQR
print("Outlier Range:", lower_bound, "to", upper_bound)

# Detect and print possible outliers
outliers = weights[(weights < lower_bound) | (weights > upper_bound)]
if len(outliers) == 0:
    print("No outliers detected.")
else:
    print("Possible outliers:", outliers)
```

# Module 2 ->  Probability Basics for ML

```python
## Part A-> basic probability

# Given data
total_people = 100
likes_pizza = 40      # A
likes_burgers = 50    # B
likes_both = 20       #A ,B both

# Probability calculations
P_A = likes_pizza / total_people
P_B = likes_burgers / total_people
P_A_and_B = likes_both / total_people

# Conditional probability P(A | B) = P(A  B) / P(B)
P_A_given_B = P_A_and_B / P_B

# Display results
print(f"P(A) = {P_A}")
print(f"P(B) = {P_B}")
print(f"P(A intersection B) = {P_A_and_B}")
print(f"P(A | B) = {P_A_given_B}")

```

```python
 ## Part B  Conditional Probability and Bayes

N = 1000
spam = 100
not_spam = N - spam

# Given
P_F_given_S = 40 / spam
P_F_given_notS = 60 / not_spam

P_S = spam / N
P_notS = not_spam / N
P_F = P_S * P_F_given_S + P_notS * P_F_given_notS
P_S_given_F = (P_S * P_F_given_S) / P_F

print("P(S | F) =", P_S_given_F)
print(f"P(S|F) = {P_S_given_F:.2f} = {P_S_given_F*100:.1f}%")

```

```python
P_D = 1 / 10000
P_pos_given_D = 0.99
P_neg_given_notD = 0.99

# Derived
P_notD = 1 - P_D
P_pos_given_notD = 1 - P_neg_given_notD

# Total probability
P_pos = P_pos_given_D * P_D + P_pos_given_notD * P_notD

# Bayes
P_D_given_pos = (P_pos_given_D * P_D) / P_pos

# Print results clearly
print(f"P(Disease) = {P_D:.6f}")
print(f"P(+ | D) = {P_pos_given_D}")
print(f"P(+ | D) = {P_pos_given_notD}")
print(f"P(+) = {P_pos:.6f}")
print(f"P(D | +) = {P_D_given_pos:.4f} = {P_D_given_pos*100:.2f}%")
```

```python
## Part C-> Confusion Matrix and Performance Metrics


# Confusion Matrix values
TP = 120   # True Positives
FN = 30    # False Negatives
FP = 80    # False Positives
TN = 770   # True Negatives

# Total samples
total = TP + FN + FP + TN

# Metrics
accuracy = (TP + TN) / total
print("Accuracy =", round(accuracy, 3))


precision = TP / (TP + FP)
print("Precisio =", round(precision, 3))

recall = TP / (TP + FN)
print("Recall =", round(recall, 3))


f1 = 2 * precision * recall / (precision + recall)
print("F1 Score =", round(f1, 3))


specificity = TN / (TN + FP)
print("Specificity =", round(specificity, 3))


npv = TN / (TN + FN)
print("NPV =", round(npv, 3))


prevalence = (TP + FN) / total
print("Prevalence   =", round(prevalence, 3))
```

1. Accuracy-> Overall correctness of the model. Here, 89% of predictions are correct.
2. Precision-> Of all predicted positives, how many are actually positive. The models precision is 60%, meaning 40% of positive predictions are wrong.
3. Recall-> Of all actual positives, how many are correctly detected. High recall (80%) means few missed positives.
4. F1 Score-> Balance between precision and recall. Here, 0.686 shows decent performance.
5. Specificity-> Ability to correctly identify negatives. High value (90.6%) means few false alarms.
6. NPV (Negative Predictive Value)-> Of all predicted negatives, how many are truly negative. Very high (96.3%) means reliable negative predictions.
7. Prevalence -> How frequent the positive class is in the dataset. Here, 15% of total cases are actually positive.

```python
import pandas as pd
# Given
total_samples = 10000
positive_rate = 0.02
positives = int(total_samples * positive_rate)
negatives = total_samples - positives


TP = 0
FN = positives
TN = negatives
FP = 0

conf_matrix = pd.DataFrame([[TP, FP],[FN, TN]],index=["Actual Positive","Actual Negative"],columns=["Predicted Positive","Predicted Negative"])

print("Confusion Matrix:")
print(conf_matrix)

# Compute metrics
Accuracy = (TP + TN) / total_samples
Precision = TP / (TP + FP) if (TP + FP) != 0 else 0
Recall = TP / (TP + FN) if (TP + FN) != 0 else 0
F1 = (2 * Precision * Recall) / (Precision + Recall) if (Precision + Recall) != 0 else 0
Specificity = TN / (TN + FP)

# Show metrics
print("\nPerformance Metrics:")
print(f"Accuracy    = {Accuracy:.4f}")
print(f"Precision   = {Precision:.4f}")
print(f"Recall      = {Recall:.4f}")
print(f"F1 Score    = {F1:.4f}")
print(f"Specificity = {Specificity:.4f}")
```
