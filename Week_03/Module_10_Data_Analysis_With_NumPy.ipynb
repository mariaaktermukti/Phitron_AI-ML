## Ndarray


```python
import numpy as np

#1D array
arr1=np.array([1,2,3,4,5])
print(type(arr1))
print(arr1.ndim)



#2D array
arr2=np.array([[1,2,3], #row 1 
               [4,5,6]]) #row 2
print(arr2.ndim)



#3D array
arr3 = np.array([
    #1st floor
    [ [1,2,3], # row1
      [4,5,6]  ], #row2 
    
    #2nd floor
     [[1,2,3], # row1
      [4,5,6]  ], #row2  


    #3rd floor
      [[1,2,3], # row1
      [4,5,6]  ] #row2  
  
])
print(arr3.ndim)
```

##Array Attributes

```python
#dimention
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)


#Shape
print(arr1.shape)      
print(arr2.shape)   
print(arr3.shape)   

#Data type
print(arr1.dtype)      
print(arr2.dtype)   
print(arr3.dtype) 


#Size
print(arr1.size)      
print(arr2.size)   
print(arr3.size) 
```

## Creating ndarray from scratch

```python
#np.zeros
arr=np.zeros((2,3), dtype=np.int8)

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)


arr = np.zeros_like(arr3)
print(arr)
print(arr.shape)
```

```python
#np.ones
arr = np.ones((3,4,3),dtype=np.int8)

print(arr) 
print(type(arr)) 
print(arr.dtype) 
print(arr.shape) 
print(arr.ndim)
print(arr.size)


arr = np.ones_like(arr3) ## arr3 shape same 
print(arr)
print(arr.shape)
```

```python
#np.empty
arr = np.empty((4,3),dtype=np.int8)

print(arr) 
print(type(arr)) 
print(arr.dtype) 
print(arr.shape) 
print(arr.ndim)
print(arr.size)

arr = np.empty_like(arr3) ## arr3 shape same 
print(arr)
print(arr.shape) 
```

### array creation with random values

```python
#np.random.rand(shape)

arr = np.random.rand(2,3) 

print(arr)
print(type(arr)) 
print(arr.dtype) 
print(arr.shape) 
print(arr.ndim)
print(arr.size)
```

```python
# np.random.randint(start,end, shape) 

arr = np.random.randint(1,10 , (2,3))  

print(arr) 
print(type(arr)) 
print(arr.dtype) 
print(arr.shape) 
print(arr.ndim)
print(arr.size)
```

```python
# np.random.uniform(start,end, shape) 

arr = np.random.uniform(50,100, (3,3,4))   

print(arr) 
print(type(arr)) 
print(arr.dtype) 
print(arr.shape) 
print(arr.ndim)
print(arr.size)
```

### array creation with range functions

```python
# np.arange(first , end , step ) -> similar to range function 

arr = np.arange(1 , 10 ,1 )
mat = arr.reshape(3,3) 

print(arr) 
print(mat) 
print(type(arr)) 
print(arr.dtype) 
print(arr.shape) 
print(arr.ndim)
print(arr.size)
```

```python
# np.linspace(start , end(included) , koita point ) 

arr = np.linspace(0, 4 ,6)


print(arr) 
print(type(arr)) 
print(arr.dtype) 
print(arr.shape) 
print(arr.ndim)
print(arr.size)
```

```python
# np.logspace(start , end(included) , koita ) 

arr = np.logspace(0 , 4 ,6, base =2) 

print(arr)
```

## indexing and slicing

```python
## indexing
print(arr1)
print(arr2)

print(arr1[2]) # 2
arr1[2]=100
print(arr1)

#2D ->  [1 2 3]
 #     [4 5 6]

print(arr2[0][2])
arr2[0][2]=100
print(arr2)
```

```python
#slicing 
#1d = arr[start:end:step] 

arr1_mod = arr1[1:4]

print(arr1_mod)
```

```python
#2D
# arr[row start : row end : step , col start : col end : step]

# print(arr2)
#getting a row
row_0= arr2[0:1, ]
print(row_0)


row_1= arr2[1: , ]
print(row_1)


#getting a column
col_0= arr2[:: , 0:1] ## value pass na korle :: dibo
print(col_0)

col_1= arr2[:: , 1:2]
print(col_1)



# getting a portion
portion= arr2[:: , 0:2]
print(portion)
```
