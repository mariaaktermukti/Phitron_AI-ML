Transpose of a Matrix


Write a Python function that computes the transpose of a given matrix.

Example:
Input:
a = [[1,2,3],[4,5,6]]
Output:
[[1,4],[2,5],[3,6]]

```python
import numpy as np

def transpose_matrix(a: list[list[int|float]]) -> list[list[int|float]]:

    a= np.array(a)
    b= a.T.tolist()

    return b
```
